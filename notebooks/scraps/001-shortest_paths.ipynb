```python
import random

def generate_route_colors(num_routes):
    color_array = ['red', 'green', 'blue', 'orange', 'yellow', 'cyan', 'magenta', 'black', 'purple', 'brown']
    if num_routes <= len(color_array):
        return random.sample(color_array, num_routes)
    else:
        return [random.choice(color_array) for _ in range(num_routes)]
```

```python
import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx

# List of coordinates
coordinates = [
    (51.51297635567389, -0.117451976654785),
    (51.519739, -0.144899),
    (51.5145, -0.1300),
    (51.5175, -0.1400)
]

# Download the street network for the specified location
# Using the average of the latitudes and longitudes to center the graph
avg_lat = sum([coord[0] for coord in coordinates]) / len(coordinates)
avg_lon = sum([coord[1] for coord in coordinates]) / len(coordinates)
G = ox.graph_from_point((avg_lat, avg_lon), dist=2000, network_type='all')

# Find the closest nodes for each set of coordinates
closest_nodes = [ox.distance.nearest_nodes(G, X=lon, Y=lat) for lat, lon in coordinates]

# Find the shortest paths between consecutive pairs of closest nodes
routes = [nx.shortest_path(G, source=closest_nodes[i], target=closest_nodes[i+1], weight='length') for i in range(len(closest_nodes)-1)]

route_colours = generate_route_colors(len(routes))

# Plot the network with all shortest paths highlighted
fig, ax = ox.plot_graph_routes(G, routes=routes, route_colors=route_colours, route_linewidth=4, show=False, close=False)

# Plot all sets of coordinates and their closest nodes
colors = ['red', 'green', 'blue', 'orange', 'yellow', 'cyan', 'magenta', 'black']  # Extend this list if you have more points
for i, (lat, lon) in enumerate(coordinates):
    ax.scatter(lon, lat, c=colors[i], s=100)
    ax.scatter(G.nodes[closest_nodes[i]]['x'], G.nodes[closest_nodes[i]]['y'], c=colors[i], s=100, edgecolors='black', linewidths=0.5)

plt.show()
```
